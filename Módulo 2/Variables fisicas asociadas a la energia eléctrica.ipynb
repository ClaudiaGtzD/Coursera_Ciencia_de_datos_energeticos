# Variables físicas asociadas a la electricidad

## Energía y potencia

**Potencia**: rapidez de la transformación de la energía

$P = \frac{\partial E}{\partial t}$

$\Delta E = \int_{}^{} P \; dt$

$\Delta E = \overline{P}X t$

Potencia: Vatios (W)

Energía: Vatios x segundos = W x s = Julio (J)

1 kWh = 3 600 000 J



## Diferencia de potencial y corriente

### Potencial eléctrico

El potencial eléctrico de un punto en el espacio es la energía eléctrica que adquiere una carga al ubcarse en ese punto, dividida por el valor de la carga (Julio / s) = Voltio (V)

### Diferencia de potencial eléctrico -- Tensión eléctrica

Es la resta de sus potenciales eléctricos

$V_{AB} = V_A - V_B$

### Corriente eléctrica

Cantidad de carga que atraviesa una superficie por unidad de tiempo

Unidades: $\frac{C}{s} = Amperio(A)$



## Relaciones entre las variables físicas

### Corriente continua

Potencia = v x i (tensión x corriente)

Energía = p x t (potencia x tiempo)

### Corriente alterna

$Potencia(t) = v(t) \times i(t)$

$Energia = \int_{}{}p(t)dt$

$v(t) = V\cos(\omega t + \varphi v)$

$\mathbf{V} = V_{\mathrm{ef}} \left| \underline{\varphi v} \right.$

$i(t) = I\cos(\omega t + \varphi i)$

$\mathbf{I} = I_{\mathrm{ef}} \left| \underline{\varphi i} \right.$

Potencia compleja

$S = VI^*$

$S = P + jQ$ La P es real y la Q es imaginaria

$E = P \times t$

## Sistemas trifásicos

### Sistema de potencia

Distribución

Transmisión

Subestaciones

Combinacion de 3 variables de tipo alterno, con misma frecuencia y misma amplitud, con desfases entre las 3 señales de 120 grados.

### Sistemas trifásicos balanceados

La potencia trifásica balanceada es constante.

Se necesitan conductores más delgados para transmitir la misma potencia.

Se disminuyen las vibraciones en las máquinas rotativas.

Menos pérdidas en conductores de la misma sección que transmiten la misma potencia.

### Sistemas trifásicos desbalanceados

Los desfases ya no está exactamente a 120 grados, hay cierta variación

**Problemas originados por el desbalance**

Aumento en las pérdidas por transmisión de energía.

Aparición de vibraciones en las máquinas rotativas.

Desgaste desigual de elementos y evenjecimiento anticipado.
