### ¿Cómo se mide la energía consumida

v(t): tensión

i(t): corriente

$\tau_{s}$: Es el intervalo de puntos en el tiempo (ms)

$p_{k}=v_{k}\times i_{k}$

$\delta E_{k}=p_{k}\times \tau_{s}$

$\Delta E=\sum_{k=1}^{T/\tau_{s}}\delta E_{k}$ La suma de energías

$P=\frac{\Delta E}{T}$ Potencia media

## Eficiencia energética

¿Cómo puede la Ciencia de Datos Energéticos hacer más sostenible nuestro entorno?

![Eficiencia](Eficiencia.PNG)

$Eficiencia = \frac{Resultados}{Insumos}$

$Eficiencia Energética = \frac{Resultados}{Energía}$

$Eficiencia Energética = \frac{Rapidez de resultados}{Potencia}$


*Por ejemplo:*

$Eficiencia Energética = \frac{Flujo luminoso}{Potencia eléctrica}$

$[Eficiencia Energética] = \frac{lúmenes}{Vatio} = \frac{lm}{W}$

Fuentes lumínicas:
- Incandescente (5-10 lm/W)
- Halógena (20-30 lm/W)
- Fluorescente (50-90 lm/W)
- LED (70-130 lm/W)

Fuentes lumínicas de productos de líneas diferente:
- Decorativo (60-80)

- Alumbrado público
- Industrial

Porque podemos obtener más resultados con la misma energía.

- Generación:

    - Balance generación-consumo:
        - Predicción del consumo
        - Predicción de la disponibilidad de las fuentes energéticas.
    - Ubicación óptima de las plantas de generación.
    - Predicciones a corto, mediano y largo plazo.
    - Planes de expansión
    - Planes de contingencia
- Transmisión:
    - Evitar la congestión de la red.
    - Flujo de carga óptimo: Ajustar la generación.
    - Predicciones a corto, mediano y largo plazo.
    - Planes de expasión
    - Planes de contingencia
- Distribución:
    - Evitar la congestion de la red.
    - Estimular un consumo más homogéneo en el tiempo.
    - Gestión de la demanda.
    - Predicciones de consumo a corto y mediano plazo.
    - Planes de expansión.
    - Planes de contingencia.
- Consumo:
    - Diseño de instalaciones más eficientes.
    - Equipos de tecnología más eficiente.
    - Uso más eficiente de los equipos.
    - Mejores hábitos de uso.
    - Autogeneración de energía.
    - Uso estratégico de las autogeneración.
    - Predicción a corto plazo de las fuentes  de autogeneración.
