# Preparación de los datos

## Análisis de componentes principales

Preparación de datos:

**Selección de características**: ignorar algunas variables.

**Reducción de la dimensionalidad**: a partir de algún procedimiento, reducimos el número de variables que contienen las variables originales.

**Selección de registros**: Seleccionar los registros con los que vamos a entrenar los datos, ignorar datos con calidad baja.

**Construcción de datos**: Crear variables nuevas que aporten al modelo (promedios, etc.)

**Integración de datos**: Combinar información de dos fuentes diferentes.

## Integración de datos

**Estandarización**

- ¿Cómo hacer comparables las tres dimensiones?
- ¿Cómo analizar simultáneamente peras y manzanas?

Llevar las variables a una sola dimensión.

*Escalador MaxMin*

$\tilde{X}_{ij}=\frac{X_{ij}-MN_{j}}{MX_{j}-MN_{j}}$

$MN_{j}=mín_{i}(X_{ij})$

$MX_{j}=máx_{i}(X_{ij})$

*Escalador Estándar*

$\tilde{X}_{ij}=\frac{X_{ij}-\mu_{j}}{\sigma_{j}}$

**Unión de tablas**

Unir tablas obtenidas de bases de datos diferentes, en base a algo que tengan en común.

## Preparación de datos de consumo de energía eléctrica

- Identificación de las variables
- Visualización de la variable de potencia, primero completo.
- Visualización de la Potencia Activa de una muestra pequeña.

Misión: construir el perfil de consumo.

- Promedio de las mediciones en una misma franja horaria, para tratarse como energía.
- Al hacer esto, se realiza una matriz donde las columnas son las horas y las filas los días.
- La suma de los renglones = consumo diario.

Al visualizar el consumo diario, se obtiene una variación mensual y semanal.

- Los promedios de las columnas de la matriz = perfil promedio horario

Al visualizar los promedios, se obtiene el perfil original medio de cada hora.

- Se puede obtener la dispersión haciendo una gráfica de caja de cada una de las horas.

Al visualizar la dispersión, se puede observar las horas con la mayor variación.
