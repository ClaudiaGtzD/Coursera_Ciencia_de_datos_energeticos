# Flujo de datos en el sector energético

## Variables eléctricas

Variables físicas medidas con algún instrumento, como las tensiones, corrientes, potencias, energía, distorciones y desfases.

**Construcción de redes inteligentes**: medidores digitales que guardan muchas variables y por medio de comunicaciones hacen llegar la información a las subestaciones.

SCADA - Supervisory Control And Data Acquisition

**Preguntas**:
¿Precisión?
¿Exactitud?
¿Confiabilidad del sistema de medición?

## Variables económicas

**Generación**: MEM y Transacciones internacionales.

**Facturación**: Precios y cantidades finales consumidas por los usuarios.

**Otros costos**: Costos de transporte, restricciones técnicas, pérdidas de energía.

**Entorno**: Inflación, monedas internacionales, mercados energéticos (Gas, Petróleo, etc.)

## Variables ambientales

### Clima y geografía

**Generación**: Hidroelecriciad, eólica, solar

**Consumo**: Calefacción, aire acondicionado

## Recursos energéticos

**Hidroelectriciad**: pluviosidad, cauces, embalses, lluvias

**Térmicas(fósil)**: disponibilidad, precio, volativilidad del precio

**Eólica**: viento, clima, estaciones

**Solar**: radiación solar, clima, estaciones

**Nuclear**: precio de combustible, amortizaciones, predicciones de precio

## Comportamiento del usuario

**Tipo de usuario**: Residencial, Industrial, Comercial

**Actividad**: Tipo de industria, aspectos demográficos y culturales

**Tamaño**: Volumen de consumo de energía

**Eventos**: Vacaciones, eventos de alcance nacional, espectáculos. Estacionalidad de la actividad

## Simulaciones del sistema eléctrico

Condiciones anormales -> Flujo de carga -> Datos sintéticos -> Análisis de seguridad

# Un marco gneral de análisis para proyectos

## 1. Dominio

¿Cuál es el dominio del sector eléctrico?

¿En qué punto de la cadena de valor de la electricidad sucede el problema?

¿Desde qué perspectiva se analiza?

## 2. Impacto efectivo

¿Qué beneficio real o esperado se obtiene con el proyecto?

¿Cómo se inserta el resultado en la cadena de valor?

## 3.Ciclo de vida

En qué etapa juega el proyecto.

Planificación -> Diseño -> Construcción -> Operación -> Mantenimiento -> Repotenciación -> Desmantelamiento

Por ejemplo: El proyecto busca mejorar la fase de mantenimiento?

## 4.Flujo de datos

¿De dónde vienen los datos?
¿Qué proceso previo han tenido los datos?

## 5. Categoría del problema

**Clasificación**: PRedecir con posibles resultados discretos.

**Regresión**: Predecir con posibles resultados numéricos.

**Agrupamiento/segmentación**: Resumir. Encontrar prototipos.

**Análisis de asociación**: Encontrar relaciones entre atributos

**Análisis de desviación**: Encontrar/Explicar comportamientos atípicos

## 6. Método de solución

Descrubrimiento de patrones

Descubrimiento de explicaciones

Descubrimiento de predictores

¿Qué tipo de métodos se han empleado?

¿Qué métodos alternativos podrían probarse?
